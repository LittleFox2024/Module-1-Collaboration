## 13.1

```python
from datetime import date
now = date.today()
fmt = "%A, %B %d, %Y"
outputfile = open("today.txt", 'wt')
print(now.strftime(fmt), file=outputfile)
outputfile.close()
```

## 15.1

```python
import multiprocessing
import random
import time

def rand_wait():
    time.sleep(random.uniform(0.0, 1.0))
    now = time.time()
    print(time.ctime(now))

if __name__ == "__main__":
    rand_wait()
    for n in range(3):
        p = multiprocessing.Process(target=rand_wait)
        p.start()
```
